```python
from sklearn import tree
import matplotlib.pyplot

from sklearn import datasets
iris = datasets.load_iris()
```

```python
print(iris.DESCR)
```

```python
print(iris.feature_names)
```

```python
x= iris.data
print("____DATA____\n",x)

y=iris.target
print("____TARGET____\n",y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix, ConfusionMatrixDisplay

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.25,random_state=25)


print('___X TEST____')
print(x_test)
print('___X TRAIN____')
print(x_train)
print('___Y TEST____')
print(y_test)
print('___Y TRAIN____')
print(x_train)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# To print accuracy_score,test class,predict class

des = MLPClassifier(hidden_layer_sizes=(30,30,20), activation='relu',)
des.fit(x_train,y_train)
y_predict=des.predict(x_test)
print("--test class--\n",y_test)
print("---predict class--\n",y_predict)
print("\n--Accuracy Score---")
print(accuracy_score(y_test,y_predict))
```

```python
cm = confusion_matrix(y_test,y_predict,normalize=None,labels=des.classes_)
print(cm)
cm_dis = ConfusionMatrixDisplay(confusion_matrix = cm,display_labels=des.classes_)
cm_dis.plot()
plt.show()
```
