```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,11)
y_square=x**2
y_cube=x**3
plt.plot(x,y_square,marker='o',linestyle='-',color='blue',label='Square')
plt.plot(x,y_cube,marker='s',linestyle='--',color='orange',label='Cube')
plt.xlabel('X-axis')
plt.ylabel('Y-Axis')
plt.title('Square and cube values')
plt.legend()
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
students=np.arange(1,11)
print(students)
physics_marks = np.random.randint(0, 101, 10)
chemistry_marks = np.random.randint(0, 101, 10)
colors = ['red','blue']
markers = ['o','*']
plt.scatter(range(10), physics_marks, c=colors[0], marker=markers[0], s=100 ,edgecolor='yellow', label='Physics Marks')
plt.scatter(range(10), chemistry_marks, c=colors[1], marker=markers[1], s=150 ,edgecolor='orange', label='Chemistry Marks')
plt.xlabel('Student')
plt.ylabel('Marks (out of 100)')
plt.title('Scatterplot of Student Marks in Physics and Chemistry')
plt.legend()
plt.show()

            

          
```

```python
a=np.arange(0,20,0.2)
b=np.sin(a)
plt.plot(a,b,linestyle='dotted',color='red')
c=np.cos(a)
plt.plot(a,c,color='green')
plt.xlabel('angle')
plt.xlabel('value')
plt.legend(['Sine','Cosine'])
plt.title('Sine and Cosine')
```
