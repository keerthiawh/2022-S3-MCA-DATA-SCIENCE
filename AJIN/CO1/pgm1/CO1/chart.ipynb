```python
from matplotlib import pyplot as plt
from matplotlib.figure import Figure
import numpy as np
x=np.arange(1,11)
y=x*x
fig=plt.figure(figsize=(6,4))
ax=fig.add_axes([1,1,1,1])
ax1=ax.plot(x,y,color='red',linewidth=2,marker="*")
z=x*x*x
ax2=ax.plot(x,z,color='green',linewidth=1,marker='.',linestyle='--')
ax.set_title("square and cube of 10 numbers")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.legend(labels=('square','cube'))
plt.grid()
```

```python
#scatterplot
import numpy as np
from numpy import random as r
import matplotlib.pyplot as plt
x=['Anu','Abhi','Manu','Nihal','Fazal','Aswani','Sreya','Varun','Fathima','Aysha']
y=[23,67,54,23,45,62,43,21,78,90]
z=[63,98,34,23,12,34,67,75,53,66]
plt.xlabel("Name of Students")
plt.scatter(x,y,color='blue',label='chemistry')
plt.ylabel("Marks Obtained")
plt.scatter(x,z,color='red',label='physics')
plt.title("Mark obtained by 10 students")
plt.legend()
plt.show()
```

```python
#histogram
import  matplotlib.pyplot as plt
x=np.random.normal(180,8,480)
plt.hist(x)
plt.show()
```
