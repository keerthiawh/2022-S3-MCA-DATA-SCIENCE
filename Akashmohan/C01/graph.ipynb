```python
#simple plot
from matplotlib import pyplot as plt
x=[5,3,4,6]
y=[7,3,10,7]
plt.plot(x)
plt.plot(y)
plt.legend(['a','b'])
plt.xlabel('X-axis Label')
plt.ylabel('y-axis Label')
plt.title('Line plot')
```

```python
#bar plot
c=['a','b','c']
values=[10,15,7]
plt.bar(c,values)
plt.xlabel('X-axis Label')
plt.ylabel('y-axis Label')
plt.title('Bar plot')
```

```python
#scatter plot
x=[1,2,3,4,5]
y=[10,12,5,8,6]
plt.scatter(x,y)
plt.xlabel('X-axis Label')
plt.ylabel('y-axis Label')
plt.title('Scatte plot')
```

```python
#histogram
import numpy as np
data=np.random.randn(1000)
plt.hist(data,bins=20)
plt.xlabel('values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# box plot
import seaborn as sns
data=sns.load_dataset('tips')
sns.boxplot(x='day',y='total_bill',data=data)
plt.xlabel('Day of the week')
plt.ylabel('Total bill')
plt.title('Box plot')
plt.show()
```

```python
x=[4,3,5]
y=[0.1,0,0]
plt.pie(x,explode=y)
plt.show()
```
