```python
import matplotlib.pyplot as plt

#squares and cube

values=[2,4,6,8,10,9,7,5,3,1]
squares=[x**2 for x in values]
cubes=[x**3 for x in values]
plt.plot(values,label='Values',color='red')
plt.plot(squares,'--',label='Squares',color='green')
plt.plot(cubes,label="Cubes")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.grid()
plt.title("Squares and Cubes")
plt.show()
```

```python
#scatterplot
import numpy as np
x=np.array([72,47,96,81,70,99,37,59,73,81])
y=np.array([29,64,69,78,92,99,87,53,39,81])
plt.plot(x,'--',label='Physics',color='green')
plt.plot(y,'.',label='Chemistry',color='pink',markersize=10,markeredgecolor='b')
plt.legend()
plt.grid()
plt.title("Marks of 10 students")
plt.show()
```

```python
#Barplot
p=['a','b','c','d','e','f','g','h','i','j']
h=[72,47,96,81,70,99,37,59,73,81]

#plt.figure(figsize=(6,6))
plt.bar(p,h,color='maroon',width=0.5)
plt.xlabel('Person')
plt.ylabel('Height')
plt.title('Bar plot')
```

```python
#Pie diagram
import numpy as np
from matplotlib import pyplot as plt
sub=['Maths','Science','Physics','Chemistry','English','History']
marks=np.array([20,67,85,79,90,76])
#h1=marks.index(max(marks))
explode=[0.1 if i == h1 else 0 for i in range(len(sub))]
colors=['blue','green','red','maroon','black','yellow']
plt.pie(marks,labels=sub,colors=colors,explode=explode)
plt.title("Pie diagram")
plt.show()
```

```python
import numpy as np
x=np.linspace(0,2*np.pi,100)
sinvalue=np.sin(x)
cosvalue=np.cos(x)
plt.plot(x,sinvalue,label='sin(x)',color="red")
plt.plot(x,cosvalue,label='cos(x)',color="green")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#Histogram

import matplotlib.pyplot as plt
x=np.random.normal(170,5,450)
plt.hist(x)
plt.title('Histogram')
plt.show()
```

```python
#Subplot

import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 1)
plt.plot(x,y)

plt.show()
```
