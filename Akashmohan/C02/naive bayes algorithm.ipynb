```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iriss=datasets.load_iris()
a=iriss.data
b=iriss.target
print(a)
print(b)
```

```python
iriss.DESCR
```

```python
a_train,a_test,b_train,b_test=train_test_split(a,b,train_size=.75,random_state=5)
print(a_train)
```

```python
print(b_train)
```

```python
print(a_train)
print(b_train)
```

```python
nb = GaussianNB ()
nb.fit(a_train,b_train)
b_predict = nb.predict(a_test)
print("-----------------------Test Class-----------------\n",b_test)
print("\n---------------------------Predict Class-----------\n",b_predict)
```

```python
print(nb.predict_proba(a_test))
```

```python
print(accuracy_score(b_test,b_predict,normalize=False))
```

```python
cm = confusion_matrix(b_test,b_predict,normalize = None)
print(cm)
```

```python
cm_dis = ConfusionMatrixDisplay(confusion_matrix=cm,)
cm_dis.plot()
plt.show()
```
