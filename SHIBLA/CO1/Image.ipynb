```python
x=ndimage.rotate(b,500,cval=20,reshape=True)
d=Image.fromarray(x)
plt.imshow(d)
```

```python
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
from PIL import Image
a=Image.open('bts.jpg')
b=np.array(a)
x=ndimage.zoom(b,1.1)
d=Image.fromarray(x)
plt.imshow(d)
```

```python
p=np.array([[1,0.5,0],[0,1,0],[0,0,1]])
x=ndimage.affine_transform(b,p)
y=Image.fromarray(x)
plt.imshow(y)
```
