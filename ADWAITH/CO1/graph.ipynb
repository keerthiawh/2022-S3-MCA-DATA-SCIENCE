```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5,6,7])
y = np.array([1,2,3,4,5,6,7])
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([10,16,18,23,25,30])
y = np.array([15,40,50,65,70,78])
plt.plot(x, y)
plt.xlabel("Age")
plt.ylabel("weight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([20, 8, 5, 30])
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([50, 30, 20, 15])
mylabels = ["cs", "eee", "mech", "civil"]
plt.pie(y, labels = mylabels)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([50, 30, 20, 15])
mylabels = ["cs", "eee", "mech", "civil"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```
