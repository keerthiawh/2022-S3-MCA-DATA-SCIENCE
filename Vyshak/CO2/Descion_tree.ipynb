```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
print(data.DESCR)
```

```python
feature_names=data.feature_names
target_names=data.target_names
```

```python
print("feature_names",feature_names)
```

```python
print("target_names",target_names)
```

```python
x=data.data
y=data.target
```

```python
#print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=46,test_size=.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion="entropy",splitter="best",max_depth=None,min_samples_split=3,min_samples_leaf=2)
from sklearn import tree
```

```python
dt.fit(x_train,y_train)
```

```python
predict=dt.predict(x_test)
```

```python
print(y_test)
print(predict)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
```

```python
print(accuracy_score(y_test,predict,normalize=True))
```

```python
cm=confusion_matrix(y_test,predict)
print(cm)
```

```python
dis=ConfusionMatrixDisplay(confusion_matrix=cm)
```
