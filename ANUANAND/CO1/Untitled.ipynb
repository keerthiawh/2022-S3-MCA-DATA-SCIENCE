```python
from matplotlib import pyplot as plt
x=[3,6,2,4]
y=[1,6,7,9]
plt.plot(x,y)
plt.show()
```

```python
#barplot
x=['A','J','S']
y=[5,8,4]
plt.bar(x,y,color='red')
plt.xlabel("Name")
plt.ylabel("Height")
```

```python
#histogram
import numpy as np
data=np.random.randn(1000)
plt.hist(data)
```

```python
#pie
x=[2,4,7]
y=[0,1,0]
plt.pie(x,explode=y)
```
