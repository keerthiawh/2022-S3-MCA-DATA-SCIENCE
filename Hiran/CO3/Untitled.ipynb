```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data=load_iris()
```

```python
print(data.DESCR)
```

```python
feature_names=data.feature_names
target_names=data.target_names
```

```python
print("feature_names :",feature_names)
print("target_names :",target_names)
```

```python
x=data.data
y=data.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=46)
```

```python
dt=DecisionTreeClassifier(criterion="entropy",splitter="best",max_depth=None,min_samples_split=4,min_samples_leaf=3)
dt.fit(x_train,y_train)
```

```python
y_predict=dt.predict(x_test)
```

```python
print("Predicted:")
print(y_predict)
print("Actual:")
print(y_test)
```

```python
print(accuracy_score(y_test,y_predict,normalize=True))
```

```python
f=plt.figure(figsize=(0,20))
tree.plot_tree(dt)
plt.show(f)
```
