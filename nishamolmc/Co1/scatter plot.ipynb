```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
physics_marks = np.random.randint(0,101,10)
chemistry_marks = np.random.randint(0,101,10)
plt.scatter(physics_marks, chemistry_marks,marker='o', label='physics marks',color='red',edgecolors='blue')
plt.xlabel('physics marks (out of 100)')
plt.ylabel('chemistry marks(out of 100)')
plt.title('scatter plot of students\'physics and chemistry marks')
plt.grid()
plt.legend()
plt.show()
```
