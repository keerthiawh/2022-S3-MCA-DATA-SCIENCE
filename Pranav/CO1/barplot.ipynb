```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Sree", "Pr", "Siv", "Shi", "Nir", "San","Ya", "Sar", "Shu", "Sah"])
y = np.array([160,180,150,155,165,160,180,170,140,190])
plt.title("Barplot")
plt.bar(x,y, color="maroon", width=0.3)
plt.xlabel('Person') 
plt.ylabel('Height')
plt.show()
```
