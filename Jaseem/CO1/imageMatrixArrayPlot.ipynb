```python
import numpy as np
from scipy import ndimage
from PIL import Image
from matplotlib import pyplot as plt
img = Image.open('pic.jpg')
Im = np.array(img)
plt.imshow(Im)
```

```python
image1=ndimage.rotate(Im, 160)
plt.imshow(image1)




```

```python


im2=ndimage.zoom(Im,(1.5,.5,.2))

plt.imshow(im2)

```
