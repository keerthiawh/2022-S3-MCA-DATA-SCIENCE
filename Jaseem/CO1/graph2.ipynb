```python
import numpy as np
from matplotlib import pyplot as plt
value= list(range(1,10))
x= [x**2 for x in value]
y=[x**3 for x in value]
plt.plot(x,color='green' ,marker = 'o',markersize = 11, )
plt.plot(y,color='red',marker = 'x',markersize = 11 )
plt.legend(["Square", "cube"])
plt.xlabel("Number")
plt.ylabel("Square and cube")
plt.grid()
```
