```python
import matplotlib.pyplot as plt
import numpy as np

names = 'Hari', 'Faheem', 'Sakthi', 'Sufail'
marks = [49.9,99.9,50,100]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.9, 0, 0, 0)  

plt.pie(marks, explode=explode, labels=names, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.show()

```
