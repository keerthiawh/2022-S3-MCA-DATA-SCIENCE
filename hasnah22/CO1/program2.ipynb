```python
#simple plot
from matplotlib import pyplot as plt
x=[5,2,4,6]
y=[7,3,10,7]
plt.plot(x,y)
plt.show()
```

```python
#bar plot
x=['anu','hasna','archana']
y=[7,4,10]
plt.bar(x,y,color='blue')
plt.xlabel("name")
plt.ylabel("marks")
```

```python
#pie
x=[2,4,7]
y=[0,1,0]
plt.pie(x,explode=y)
```
