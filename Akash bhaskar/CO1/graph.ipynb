```python
import numpy as np
from matplotlib import pyplot as plt
values=[1,3,4,5,6,7,8,3,4,0]
square=[x**2 for x in values]
cube=[x**3 for x in values]
plt.plot(values,label=values)
plt.plot(square,'--',label=square)
plt.plot(cube,label=cube)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Square and cube ')
plt.legend()
plt.show()
```

```python
#pie diagram
import numpy as np
from matplotlib import pyplot as plt
x=['java','c','ase','php','dbms','ai']
y=np.array([75,44,104,34,77,66])
# a=y.index(max(y))
explode=[0.1 if i==a else 0 for i in range(len(x))]
colors=['cyan','purple','yellow','pink','black','red']
plt.pie(y,labels=x,colors=colors,explode=explode)
plt.title('PIE DIAGRAM')
plt.show()
```

```python
#scatterplot
x=np.array([72,47,96,81,70,99,37,59,73,81])
y=np.array([29,64,69,78,92,99,87,53,39,81])
plt.plot(x,'--',label='Physics',color='green')
plt.plot(y,'.',label='Chemistry',color='pink',markersize=10,markeredgecolor='b')
plt.legend()
plt.grid()
plt.title("Marks of 10 students")
plt.show()
```

```python
#Barplot
p=['a','b','c','d','e','f','g','h','i','j']
h=[72,47,96,81,70,99,37,59,73,81]

#plt.figure(figsize=(6,6))
plt.bar(p,h,color='maroon',width=0.5)
plt.xlabel('Person')
plt.ylabel('Height')
plt.title('Bar plot')
```

```python
import numpy as np
x=np.linspace(0,2*np.pi,100)
sinvalue=np.sin(x)
cosvalue=np.cos(x)
plt.plot(x,sinvalue,label='sin(x)',color="red")
plt.plot(x,cosvalue,label='cos(x)',color="green")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#sub plot
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.plot(x,y)
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.show()
```
