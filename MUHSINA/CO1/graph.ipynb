```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([2,6,5])
y=np.array([3,9,4])

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt 

x1 = [1,2,3] 
y1 = [2,4,1] 
plt.plot(x1, y1, label = "line 1") 

x2 = [1,2,3] 
y2 = [4,1,3] 
plt.plot(x2, y2, label = "line 2") 

plt.xlabel('x - axis') 
plt.ylabel('y - axis') 
plt.title('Two lines on same graph!') 

plt.legend() 

plt.show() 
```

```python
import matplotlib.pyplot as plt 

x = [1,2,3,4,5,6] 
y = [2,1,8,7,12,9] 

plt.plot(x, y, color='red', linestyle='dashed', linewidth = 3, 
		marker='o', markerfacecolor='black', markersize=12) 

plt.ylim(1,12) 
plt.xlim(1,12) 

plt.xlabel('x - axis') 
plt.ylabel('y - axis') 

plt.title('Some cool customizations!') 

plt.show() 
```

```python
import matplotlib.pyplot as plt 


left = [1, 2, 3, 4, 5] 

height = [10, 24, 36, 40, 5] 

tick_label = ['A', 'B', 'C', 'D', 'E'] 

plt.bar(left, height, tick_label = tick_label, 
		width = 0.8, color = ['yellow', 'blue']) 

plt.xlabel('x - axis') 
plt.ylabel('y - axis') 
plt.title('My bar chart!') 

plt.show() 
```
