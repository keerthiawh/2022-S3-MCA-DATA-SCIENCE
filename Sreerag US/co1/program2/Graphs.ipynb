```python
#simple plot
import numpy as np
from matplotlib import pyplot as plt
x=[5,2,4,6]
y=[7,3,10,7]

plt.plot(x,marker='*')
plt.plot(y)
plt.legend(['first line','second line'])
plt.show()
```

```python
#bar plot
x=['Anu','sweta','Aysha']
y=[7,4,10]
plt.bar(x,y,color='green')
plt.xlabel("name")
plt.ylabel("cgpa")
plt.title("cgpa of students")
```

```python
#histogram plot
x=np.random.rand(1000)
plt.hist(x,bins=20,color='yellow')
plt.show()
```

```python
#pie plot
x=[2,4,7]
y=[0.1,0,0]
plt.pie(x,explode=y)
plt.legend(['apple','orange','grape'])
```

```python
x=[3,4,6,1]
y=[4,7,1,7]
plt.scatter(x,y,marker='*')
```

```python
#sine
x=np.arange(0,20,0.1)
y=np.sin(x)
plt.plot(x,y)
```

```python
#cosine
x=np.arange(0,20,0.1)
y=np.cos(x)
plt.plot(x,y)
```

```python
a=np.arange(0,20,0.1)
b=np.cos(a)
c=np.arange(0,20,0.1)
d=np.sin(c)
plt.subplot(1,3,3)
plt.plot(a,b,c,d)
```
