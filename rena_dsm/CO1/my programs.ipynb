```python
import cv2
import numpy as np
a=Image.open('rose flower.jpg')
rows,cols,_ = a.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(a,M,(cols,rows))
cv2.imshow("Affine Transform", dst)
```
