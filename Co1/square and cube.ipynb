```python
import numpy as np
import matplotlib.pyplot as plt
x= np.linspace(1,10,10)
y_square = x**2
y_cube = x**3
plt.figure(figsize=(8,6))
plt.plot(x,y_square,label='square', color='green' , marker='*')
plt.plot(x,y_cube, label='cube', color='red',linestyle='dashed', marker='.')
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('square asnd cube of 10 values')
plt.legend()
plt.grid()
plt.show()
```
