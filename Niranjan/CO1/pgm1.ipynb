```python
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
from PIL import Image
a=Image.open('images.jpg')
plt.imshow(a)
```

```python
b=np.array(a)
x=ndimage.rotate(b,90)
c=Image.fromarray(x)
plt.imshow(c)
```

```python
z=ndimage.zoom(b,(1.2,1.2,1))
d=Image.fromarray(z)
plt.imshow(d)
```

```python
p=Image.open('abc.jpg')
mat=np.array([[1,0,0],[0,1,0],[0,0,1]])
mat
im=ndimage.affine_transform(p,mat)
plt.show(im)
```
