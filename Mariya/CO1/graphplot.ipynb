```python
import matplotlib.pyplot as plt 
x=[1,2,3,4,5,6]
y=[2,4,3,6,7,3]
plt.plot(x, y,linewidth = 2, label = "line 1")
x1=[1,2,3,4,5,6]
y1=[3,5,4,3,2,4]
plt.plot(x1, y1,color='green',linestyle='dashed',linewidth = 2, label = "line 2")
plt.xlabel('x - axis') 
plt.ylabel('y - axis') 
plt.title('My graph!')
plt.legend()
plt.show()
```
