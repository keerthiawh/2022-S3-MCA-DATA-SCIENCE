```python
import numpy as np
from PIL import Image
from scipy import ndimage
from matplotlib import pyplot as plt
a=Image.open('image.jpg')
img_array=np.array(a)
rotate_array=np.rot90(img_array,k=2)
rotated_img=Image.fromarray(rotate_array)
plt.imshow(rotated_img)
```

```python
y=ndimage.zoom(img_array,(.25,.25,1))
w=Image.fromarray(y)
plt.imshow(w)
```
