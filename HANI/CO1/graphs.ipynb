```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,11)
y_square=x**2
y_cube=x**3
plt.plot(x,y_square,marker='o',linestyle='-', color='blue',label='Square')
plt.plot(x,y_cube,marker='s',linestyle='--', color='orange',label='Cube')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Square and cube values')
plt.legend()
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
students=np.arange(1,11)
p_marks=np.random.randint(0,101,size=10)
c_marks=np.random.randint(0,101,size=10)
plt.scatter(students,p_marks,color='blue',edgecolors='black', s=100, marker='o',label='Physics')
plt.scatter(students,c_marks,color='green',edgecolors='black', s=80, marker='s',label='Chemistry')
plt.xlabel('Students')
plt.ylabel('Marks (out of 100)')
plt.title('Physics and chemistry marks - Scatter Plot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
persons=np.arange(1,11)
heights=np.random.randint(140,190,size=10)
plt.bar(persons, heights, color='maroon',width=0.5, label='Height')
plt.xlabel('Persons')
plt.ylabel('Height (cm)')
plt.title('Height of 10 persons - bar plot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
marks=[100,60,80,90,75,70]
exp=[0.1,0,0,0,0,0]
col=['pink','red','blue','green','yellow','brown']
labels=['or','daa','dsa','sna','dsm','mad']
plt.pie(marks,explode=exp,colors=col,labels=labels)
plt.title('PIE DIAGRAM')
```

```python
a=np.arange(0,20,0.2)
b=np.sin(a)
plt.plot(a,b,linestyle='dotted',color='red')
c=np.cos(a)
plt.plot(a,c,color='green')
plt.xlabel('angle')
plt.xlabel('value')
plt.legend(['sine','cosine'])
plt.title('Sine and cosine')
```
