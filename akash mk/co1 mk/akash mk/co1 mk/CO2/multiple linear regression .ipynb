```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris=datasets.load_iris()
x_values=iris.data[:,0]
y=iris.target
print(x_values)
print(y)
```

```python
print(iris.DESCR)
```

```python
print(iris.feature_names)
```

```python
print(iris.target_names)
```

```python
x=iris.data
print(x.shape)
print(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=.25,random_state=1)
print("----------x_train--------")
print(x_train)
print("----------x_test---------")
print(x_test)
print("----------y_train--------")
print(y_train)
print("----------y_test---------")
print(y_test)
```

```python
sl_reg=LinearRegression()
sl_reg.fit(x_train,y_train)
y_predict=sl_reg.predict(x_test)
print("-----test class--------\n",y_test)
print("----------predict class-----\n",y_predict)
```

```python
print(sl_reg.coef_)
print(sl_reg.intercept_)
```
