```python
#simple plot
from matplotlib import pyplot as plt
x=[5,2,4,6]
y=[7,3,10,7]
plt.plot(x,y)
plt.show()
```

```python
#bar plot
x=['akash','sufail','nisha']
y=[7,4,10]
plt.bar(x,y,color='green')
plt.xlabel("name")
plt.ylabel("mark")
```

```python
#histogram plot
x=[2,4,6,8,12,20,31,23,13,3]
plt.hist(x)
```
