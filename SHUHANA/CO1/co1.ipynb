```python
import numpy as np
from matplotlib import pyplot as plt
che= np.array([25,70,70,45,78,28,80,34,27,90])
phy= np.array([65,60,80,45,89,28,20,34,47,34])
plt.scatter(che,phy, color='red',marker='*',label='marks',edgecolor='yellow',s=250)
plt.xlabel('che')
plt.ylabel('phy')
plt.title('students mark')
plt.legend()
plt.show()


```

```python
height =[162,182,132,170,184,121,106,154,160,140]
plt.bar(range(10),height,color='maroon',width=0.5,align='center')
plt.xlabel('persons')
plt.ylabel('height')
plt.title('height of 10 persons')
plt.show()
```

```python
x = np.random.normal(170,2,200)

plt.hist(x,color='violet')
plt.show()
```

```python
subject = ["Arabic", "social", "Chemistry", "physics","biology","english"]
marks=[50,25,55,77,38,44]
heighest_marks=marks.index(max(marks))
explode=[0] *len(subject)
explode[heighest_marks]=0.1
plt.pie(marks,labels=subject,startangle=90,explode=explode)
plt.title('students marks')
plt.show()
```

```python
x = np.arange(0, 5*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, color='green',linestyle='dotted')
plt.plot(x, y2, color='darkblue',linestyle='solid')
plt.xlabel('pi value')
plt.ylabel("sinx cosx")
plt.title('data')
plt.legend(['sin(x)','cos(x)'])
plt.show()
```

```python
data = np.random.normal(30,2,100)
plt.boxplot(data)
plt.xlabel('value')
plt.title('quartile plot')
plt.show()
```
