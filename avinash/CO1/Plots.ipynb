```python
import matplotlib.pyplot as plt
x=list(range (1,11))
squares=[val ** 2 for val in x]
cube=[val ** 3 for val in x]
plt.figure(figsize=(8,6))

plt.plot(x,squares,label="squares",linestyle="dotted",marker="*",color="brown")
plt.plot(x,cube,label="cube")
plt.xlabel("x-xaxis")
plt.ylabel("y-axis")
plt.title("squares and cube")
plt.legend()
plt.grid()
plt.show()
```

```python
#scatterplot
import numpy as np
import matplotlib.pyplot as plt
name=['avinash','abhi',"arun",'varun','arjun','sarun','abhay','akash','vaishnav','hani']
physics_marks=np.array([50,100,10,75,30,60,58,80,66,100])
chemistry_marks=np.array([89,90,76,60,50,45,55,67,89,67])
plt.scatter(name,physics_marks,label="physics",color="red",marker="H",edgecolors="black",s=100)
plt.scatter(name,chemistry_marks,label="chemistry",color="yellow",marker="P",edgecolors="black",s=100)
plt.xlabel("physics marks")
plt.ylabel("chemistry marks")
plt.title("physics vs chemistry marks for 10 students")
plt.legend()
plt.show()
```

```python
#bar plot
import matplotlib.pyplot as plt
name=['avinash','abhi',"arun",'varun','arjun','sarun','abhay','akash','vaishnav','hani']
heights=[175,161,155,133,132,167,188,189,156,145]
plt.bar(name,heights,color="maroon",width=0.5,label="heights in cm")
plt.xlabel("person")
plt.ylabel("heigth (cm)")
plt.title("heights of 10 persons")
plt.legend() 
plt.show()
```

```python
#histogram
d=np.random.normal(170,2,200)
plt.hist(d,color="blue",edgecolor="black")
plt.title("histogram of heights")
```

```python
#pie plot
marks=[100,60,40,90,45,78]
exp=[0.1,0,0,0,0,0]
mylabels=['maths','physics','chemistry','english','social','hindi']
col=["pink","red","blue","brown","yellow","green"]

plt.pie(marks,explode=exp,colors=col,labels=mylabels,shadow=True)

plt.show()
```

```python
#sine cosine
x=np.arange(0,20,0.2)
y=np.sin(x)
plt.plot(x,y,linestyle="dotted",color="red")
c=np.cos(x)
plt.plot(x,c,color="green")
plt.xlabel('angle')
plt.ylabel('value')
plt.legend(["sine","cosine"])
plt.title("sine and cosine")
```
