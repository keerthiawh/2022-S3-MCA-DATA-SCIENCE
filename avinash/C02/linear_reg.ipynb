```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
import numpy as np
```

```python
iris=datasets.load_iris()
x=iris.data
print(x)
```

```python
print(iris.DESCR)
```

```python
y=iris.target
print(y)
```

```python
print(iris.feature_names)
```

```python
print(iris.target_names)
```

```python
print("dataset")
z=x.reshape(-1,1)
z=z[0:150]
print(z)
```

```python
x_train,x_test,y_train,y_test=train_test_split(z,y,test_size=.25,random_state=4)
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
y_predict=reg.predict(x_test)
print("---------------test_class-------------\n",y_test)
print("---------------predict_class---------------\n",y_predict)
print("coeffiecient:",reg.coef_)
print("intercept:",reg.intercept_)
```

```python
err=mean_squared_error(y_test,y_predict)
print(err)
rsc=r2_score(y_predict,y_test)
print(rsc)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_predict)
```
