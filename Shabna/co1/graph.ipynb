```python
import numpy as np
from matplotlib import pyplot as plt
x=[5,2,4,6]
y=[7,3,10,7]
plt.plot(x,marker='*')
plt.plot(y)
plt.legend(['first line','second line'])
plt.show()

```

```python
x=['Anu','sweta','Aysha']
y=[7,4,10]
plt.bar(x,y,color='green')
plt.xlabel("name")
plt.ylabel("cgpa")
plt.title("cgpa of students")
```

```python
x=np.random.rand(1000)
plt.hist(x,bins=20,color='yellow')
plt.show()
```

```python
x=[2,4,7]
y=[0.1,0,0]
plt.pie(x,explode=y)
plt.legend(['apple','orange','grape'])
```
