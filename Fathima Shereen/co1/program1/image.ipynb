```python
import numpy as np
from scipy import ndimage
from PIL import Image
from matplotlib import pyplot as plt
a=Image.open('cat.jpg')
b=np.array(a)
plt.imshow(a)
```

```python
#rotate
x=ndimage.rotate(b,90)
m=Image.fromarray(x)
plt.imshow(m)
```

```python
#zoom
y=ndimage.zoom(b,(0.4,0.4,1))
n=Image.fromarray(y)
plt.imshow(n)
```

```python
#affine_transform
z=ndimage.affine_transform(b)
```
