```python
import numpy as np
from PIL import Image
from scipy import ndimage
from matplotlib import pyplot as plt
a=Image.open('dog.jpg')
plt.imshow(a)
```

```python
b=np.array(a)
c=ndimage.rotate(b,180)
plt.imshow(c)
```

```python
z=ndimage.zoom(b,(0.4,0.4,1))
d=Image.fromarray(z)
plt.imshow(d)
```

```python
p=np.array([[1,0.5,0],[0,1,0],[0,0,1]])
x=ndimage.affine_transform(b,p)
y=Image.fromarray(x)
plt.imshow(y)
```
